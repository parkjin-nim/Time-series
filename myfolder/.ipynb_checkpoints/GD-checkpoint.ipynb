# GD

```python
import matplotlib.pyplot as plt
import numpy as np

def obj_f(x):
    return x**4

def gd(x,
      it=200):
    x_hist = []
    y_hist = []
    for n in range(100):
        x_hist.append(x)
        y_hist.append(obj_f(x))

        x = x - 0.01 * 4*x**3

    return x_hist, y_hist

x_out,y_out = gd(2.7)

# plot final generation fitness
x_input =np.linspace(start=-3, stop=3, num=200)
plt.plot(x_input, obj_f(x_input))
# plot best parents fitness among them.
plt.scatter(x_out, y_out, marker='.', c='b', s=200)
plt.pause(0.09)
plt.ioff()
```

# Newton's method

```python
def nt(x,
      it=200):
    x_hist = []
    y_hist = []
    for n in range(it):
        x_hist.append(x)
        y_hist.append(obj_f(x))

        x = x - (x**4)/(4*x**3)

    return x_hist, y_hist

x_out,y_out = nt(2.7, 20)

# plot final generation fitness
x_input =np.linspace(start=-3, stop=3, num=200)
plt.plot(x_input, obj_f(x_input))
# plot best parents fitness among them.
plt.scatter(x_out, y_out, marker='.', c='b', s=200)
plt.pause(0.09)
plt.ioff()
```

# Gauss Newton

```python
def gnt(x,
      it=200):
    x_hist = []
    y_hist = []
    for n in range(it):
        x_hist.append(x)
        y_hist.append(obj_f(x))

        x = x - (4*x**3)/(12*x**2)

    return x_hist, y_hist

x_out,y_out = gnt(2.7, 20)

# plot final generation fitness
x_input =np.linspace(start=-3, stop=3, num=200)
plt.plot(x_input, obj_f(x_input))
# plot best parents fitness among them.
plt.scatter(x_out, y_out, marker='.', c='b', s=200)
plt.pause(0.09)
plt.ioff()
```

# Levenberg-Marqardt

```python
def gnt(x,
      it=200):
    x_hist = []
    y_hist = []
    for n in range(it):
        x_hist.append(x)
        y_hist.append(obj_f(x))

        x = x - (4*x**3)/(abs(12*x**2) + 1000)

    return x_hist, y_hist

x_out,y_out = gnt(2.7, 20)

# plot final generation fitness
x_input =np.linspace(start=-3, stop=3, num=200)
plt.plot(x_input, obj_f(x_input))
# plot best parents fitness among them.
plt.scatter(x_out, y_out, marker='.', c='b', s=200)
plt.pause(0.09)
plt.ioff()
```
